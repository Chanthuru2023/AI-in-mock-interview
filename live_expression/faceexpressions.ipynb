```python
import cv2
import numpy as np
import urllib
import matplotlib.pyplot as plt
from deepface import DeepFace

# URL of an image (you can replace it with any image URL)
url = "https://as1.ftcdn.net/v2/jpg/05/03/20/60/1000_F_503206090_1W6w9s9kSVqZZGZX76ThWFZCDiGMxtMr.jpg"

# Load the image from the URL
req = urllib.request.urlopen(url)
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
sample_image = cv2.imdecode(arr, -1)

# Analyze the facial expression using DeepFace
analysis_result = DeepFace.analyze(sample_image, actions=['emotion'], enforce_detection=False)

# Since the result is a list, we need to access the first face's emotion analysis
emotion_result = analysis_result[0]['emotion']

# Display the image and analysis results
plt.imshow(cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide axis
plt.show()

# Print emotion analysis result for the first detected face
print("Emotion Analysis Result: ", emotion_result)
```
